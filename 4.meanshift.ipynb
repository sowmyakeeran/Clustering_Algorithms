                                  4. mean shift clustering
    It is a centroid based algorithm.It works by shifting data points towards centroids
    to be mean of other data points in the region

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset.head()
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
plt.scatter(dataset['Annual Income (k$)'],dataset['Spending Score (1-100)'])
plt.xlabel('income')
plt.ylabel('spending score')
```

```python
from sklearn.cluster import MeanShift
clustering = MeanShift(bandwidth=2)
labels=clustering.fit_predict(x)
```

```python
supervised=pd.DataFrame(dataset)
supervised['clusters']=labels
supervised.head()
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```

```python
#evalution metrics for clustering#silhoutte_score
from sklearn.metrics import silhouette_score
silhouette_score(x,labels)
```
